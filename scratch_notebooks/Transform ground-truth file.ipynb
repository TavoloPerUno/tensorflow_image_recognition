```python
import pandas as pd
import os
from skimage import io
```

```python
df_data = pd.read_csv(os.path.join('..', 'data', 'prediction', 'all_cities_manual_image_recognition.csv') )
```

```python
df_data.columns
```

```python
df_data = df_data[['id', 'x', 'Ped', 'Cyclists', 'Parked_cycles',
       'Cars', 'Buses', 'Motorbikes', 'Vans']]
```

```python
df_data.columns = ['id', 'location', 'pedestrian', 'bicyclist', 'parked_bicycle',
       'car', 'bus', 'motorbike', 'van']
```

```python
df_data = pd.melt(df_data, id_vars=['id', 'location'], value_vars=['pedestrian', 'bicyclist', 'parked_bicycle',
       'car', 'bus', 'motorbike', 'van'])
```

```python
df_data.columns = ['id', 'location', 'class', 'count']
```

```python
df_data['count'][df_data['count'] < 0] = 0
```

```python
df_data.to_csv(os.path.join('..', 'data', 'prediction', 'ground_truth.csv'), header=True, index=False)
```

```python
for idx, row in df_data.loc[df_data['id'] > 10016,][['id', 'location']].drop_duplicates().iterrows():
    try:
        output_img = os.path.join('..', 'data', 'prediction', 'groundtruth_images', str(row['id']) + '.jpg')
        img = io.imread(row['location'])
        io.imsave(output_img, img)
    except Exception as ex:
        print(ex)
        print("Error process image " + str(row['id']))
```
